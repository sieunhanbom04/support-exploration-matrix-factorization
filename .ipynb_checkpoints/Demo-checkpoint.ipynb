# 1. Bilinear Hard Thresholding Pursuit (BHTP) for MEG factorization

```python
from support_explore import BHTP
import numpy as np
import matplotlib.pyplot as plt
from pyfaust.factparams import ParamsPalm4MSA, StoppingCriterion
from pyfaust.fact import palm4msa, palm4msa_mhtp
from pyfaust.proj import splin, spcol
from pyfaust.factparams import MHTPParams
from time import time
```

File **support_explore.py** is the original one used to produce the Figure 1 in the paper. We use the function **BHTP** to perform the factorization of the matrix MEG.

```python
# load MEG matrix
MEG = np.load("MEG.npy")

# number of coefficients per row and per column of factors in the paper
k0 = 25
k = [25, 50, 75, 100]
loss_PALM = []
loss_BHTP = []

for k1 in k:
    # PALM (BHTP without explicitly changing of the support)
    print("k0 = ", k0, " k1 = ", k1)
    print("PALM")
    evol1 = BHTP(MEG, k0, k1, max_iter = 100, flag = False)
    loss_PALM.append(np.min(evol1))
    
    # BHTP
    print("BHTP")
    evol2 = BHTP(MEG, k0, k1, max_iter = 100, flag = True)
    loss_BHTP.append(np.min(evol2))
```

```python
plt.plot(k, loss_PALM, marker='o', label = 'PALM', color = 'blue', markersize=10)
plt.plot(k, loss_BHTP, marker='o', label = 'BHTP', color = 'red', markersize=10)
plt.legend(prop={'size' : 15})
plt.xlabel('k1', fontsize = 15)
plt.ylabel('Relative error', fontsize = 15)
plt.xticks(k, fontsize = 15)
plt.yticks(np.linspace(0, 0.18, 10), fontsize = 15)
```

Alternatively, we have another version of BHTP implemented in the library Fa$\mu$st. Its name is MHTP (standing for Multi-linear Hard Thresholding Pursuit). This version a finer one written in C++. The following code is adapted from <a href="https://faustgrp.gitlabpages.inria.fr/faust/last-doc/html/MHTP_algorithms.html">Fa$\mu$st library documentation</a> (more detail can be also found there).

```python
# load MEG matrix
MEG = np.load("MEG.npy").astype('float32')

# number of coefficients per row and per column of factors in the paper
k0 = 25
k1 = 25

# projection corresponding to each factors
projs = [splin((8193, 204), k0), spcol((204, 204), k1)]
# the number of iterations of PALM4MSA
stop_crit = StoppingCriterion(num_its=2000)
param = ParamsPalm4MSA(projs, stop_crit, [np.eye(MEG.shape[0], MEG.shape[1]), np.eye(MEG.shape[1])], 
                       is_verbose=False, no_lambda = True, no_normalization = True)
param.factor_format = 'dense' # working with dense matrices

# parameters for BHTP
mhtp_p = MHTPParams(updating_lambda = False)
print(mhtp_p)
```

```python
F_palm4msa = palm4msa(MEG, param, backend=2020)
err_palm4msa = (F_palm4msa-MEG).norm()/np.linalg.norm(MEG)
print("PALM4MSA error:", err_palm4msa)
```

```python
F_mhtp = palm4msa_mhtp(MEG, param, mhtp_p)
err_palm4msa_mhtp = (F_mhtp-MEG).norm()/np.linalg.norm(MEG)
print("PALM4MSA-MHTP error:", err_palm4msa_mhtp)
```

# 2. Hadamard factorization with $k$-sparse matrices

In this section, we will apply the projection demonstrated in the file **k_assignment.py** for the factorizatoin of the Hadamard factorization. However, we will not use the code directly since it takes very long time for the projection. This is not the problem of the complexity of the algorithm, but rather due to the implemented language. As you can see in the file **k_assignment.py**, the code contains main *while* and *for* loops, which takes a lot of time in python. Therefore, we will use a C++ implementation instead. It is wrapped under the library *pyfaust*.

```python
from pyfaust.proj import skperm, splincol
from k_assignment import Kcolrow
import time
from palm4msa_hierar import palm4msa_py, hierarchical_py
import palm4msa_hierar2
from pyfaust.factparams import *
from pyfaust import *
```

To show you the difference between the performance, let check the following example

```python
size = 256
faust_proj = skperm((size, size), 2) # projection implemented in C++ of Faust
assignment_proj = Kcolrow((size, size), 2) # projection implemented purely in Python (code in k_assignment.py)

# random a matrix to project
A = np.random.randn(size, size)

# Faust projection
begin = time.time()
faust_A_proj = faust_proj(A)
end = time.time()
print("Running time of faust projection: ", end - begin)

# k_assignment.py projection
begin = time.time()
assignment_A_proj = assignment_proj(A)
end = time.time()
print("Running time of k_assignment projection: ", end - begin)

# Compare two results
print("Difference between two projections: ", np.linalg.norm(faust_A_proj - assignment_A_proj))
```

We emphasize again that both implementation are identical (algorithmic-wise). However, an implementation in C++ can outperform largely its counterpart in Python. The rest of this section will use **skperm** - a C++ version of **Kcolrow** defined in **k_assignment.py**.

In the rest of this notebook, we show how to factorize the Hadamard transform with skperm and PALM and hierarchical factorization.

```python
factor = 9
size = 2 ** factor
H = wht(size).toarray()
N = 30
```

```python
# result using old projection splincol
proxs = [splincol((size,size), 2) for i in range(factor)]
F1, _lambda = palm4msa_py(H, factor, N, proxs, S = 'zero_and_ids', is_update_way_R2L=True)
print("Error approximation: ", np.linalg.norm(F1.toarray() - H))
```

```python
# result using skperm
proxs = [skperm((size,size), 2) for i in range(factor)]
F1, _lambda = palm4msa_py(H, factor, N, proxs, S = 'zero_and_ids', is_update_way_R2L=True)
print("Error approximation: ", np.linalg.norm(F1.toarray() - H))
```

Alternatively, you can use **Kcolrow** to perform the factorization. However, we strongly recommend you to use it with small case since it can take a lot of time

```python
# result using Kcolrow
proxs = [Kcolrow((size,size), 2) for i in range(factor)]
F1, _lambda = palm4msa_py(H, factor, N, proxs, S = 'zero_and_ids', is_update_way_R2L=True)
print("Error approximation: ", np.linalg.norm(F1.toarray() - H))
```

Finally, you can use **skperm** for left factor in hierarchical factorization to factorize the Hadamard transform. Note that hierarchical factorization is costlier than PALM.

```python
# result using skperm for the left factors of hierarchical factorization
F1 = hierarchical_py(H, J=factor, N=30, fac_proxs=[skperm((size,size), 2) for i in range(factor - 1)],
                    res_proxs=[splincol((size,size), int(size/2**(i+1))) for i in range(factor - 1)],
                    is_update_way_R2L=True, is_fact_side_left=False)
print("Error approximation: ", np.linalg.norm(F1.toarray() - H))
```

As you can see, **skperm** allows both PALM and hierarchical factorization to exactly recover the Hadamard transform (we mean the difference $\|A - X_1 \ldots X_n\|$ is of order $10^{-14}$. In constrast, this is not the case for the old projection.
